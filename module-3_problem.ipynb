# Problem-1

```python
# File read write from .txt
# 1. Number of line
# 2. Number of words
# 3. Number of character
# and save it to the conter_of_string.txt file

from functools import reduce
with open("./sample_data/sample.txt","r") as file:
    strings = file.readlines()

    # Total lines
    total_lines = len(strings)

    # Total words
    number_of_words = list(map(lambda x: len(x.split()), strings))
    
    total_words = reduce(lambda x,y : x+y, number_of_words)

    # Cleaning process
    # 1. new line delet
    strings = list(map(str.strip, strings))

    # 2. space delete
    strings = list(map(lambda x : x.replace(" ", ""), strings))

    # Characters
    number_of_characters = list(map(lambda x : len(x), strings))
    print(number_of_characters)
    
    # Total characters
    total_characters = reduce(lambda x,y : x+y, number_of_characters)
    print(total_characters)

    with open("./sample_data/counter_of_string.txt", "w") as file :
        file.write(f"Total lines --> {total_lines} \nTotal words --> {total_words} \nTotal characters --> {total_characters}")
    
```

# Problem-2

```python
# Write a function create_file(filename, content) that creates a new text file with the given filename and writes the content into it. Call it using positional arguments.

def create_file(filename, contant):
    with open(filename, "w") as file :
        file.write(contant)

create_file ("notes.txt", "Hello, this is my first file")
```

# Problem-3

```python
# Write a function read_lines(filename) that reads a text file and prints each line with its line number.

def read_line(filename):
    with open(filename, "r") as file:
        count = 1
        for line in file:
            print(f"{count}. {line}", end = "")

            count += 1

read_line("notes.txt")
```

```python
# Given a filename log.txt, use file append mode to add a new log entry each time the function add_log(message) is called. Each message should appear on a new line.

def add_log(message):
    with open("log.txt", "a") as file:
        file.write(message + "\n")

add_log("User logged in")
add_log("File uploaded")
    
```

# Problem-5

```python
# Write a function read_from_position(filename, position) that uses the file pointer to start reading a file from a given byte position and prints the remaining content.

def read_from_position(filename, position):
    with open(filename, "r") as file:
        file.seek(position)

        content = (file.read())

        print(content)

read_from_position("notes.txt", 24)
```

# Problem_6

```python
"""Write a function safe_read(filename) that tries to open and read a file. Use try, except, else, and finally blocks:
Print file content if successful (else block)
Handle the case where file does not exist (except block)
Always print "Operation complete." at the end (finally block)
"""

def safe_read(filename):
    try:
        with open(filename, "r") as file:
            content = file.read()
    except FileNotFoundError:
        print("---File was not found---")

    else:
        print(content)
    finally:
        print("Operation complete")

safe_read("data.txt")
```
